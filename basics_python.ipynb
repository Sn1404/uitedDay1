```python
class Person:
    def __init__(self,name,dept):
        self.name=name
        self.dept =dept
    def get_Details(self):
        return f"Name:{self.name} DEPARTMENT:{self.dept}"
p1 =Person("Nitish","CS")
print(p1.name)
print(p1.dept)
print(p1.get_Details())
```
